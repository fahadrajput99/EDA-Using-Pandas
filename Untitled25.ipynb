```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/homeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
```

```python
dummies
```

```python
final=pd.concat([df,dummies],axis=1)
```

```python
dummies
```

```python
final=final.drop(['town','west windsor'],axis=1)
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop(['price'],axis=1)
```

```python
y=final.price
```

```python
x
```

```python
y
```

```python
model.fit(x,y)
```

```python
model.predict([[3000,1,0]])
```

```python
model.predict([[2800,0,1]])
```

```python
model.score(x,y)*100
```
