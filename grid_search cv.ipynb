```python
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
df['Species']=la.fit_transform(df['Species'])
```

```python
df
```

```python
y=df.Species
input=df.drop(['Species'],axis=1,inplace=True)
```

```python
input
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test =train_test_split(df,y,test_size=0.2)
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', C=30, gamma='auto')
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
```

```python
accuracy
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear', C=10, gamma='auto'),x_train,y_train,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', C=20, gamma='auto'),x_train,y_train,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', C=30, gamma='auto'),x_train,y_train,cv=5)
```

```python
from sklearn.linear_model import LogisticRegression
la=LogisticRegression()
la.fit(x_train,y_train)
accuracy=la.score(x_test,y_test)
```

```python
accuracy
```

```python
import numpy as np
kernals=['rbf','linear']
c=[1,10,20]
avg_score={}
for kval in kernals:
  for cval in c:
    cv_score=cross_val_score(SVC(kernel='rbf', C=30, gamma='auto'),x_train,y_train,cv=5)
    avg_score[kval + '_' + str(cval)]=np.average(cv_score)

avg_score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

parameters = {
    'kernel': ['rbf', 'linear'],
    'C': [1, 10, 20],
}

clf = GridSearchCV(SVC(gamma='auto'), parameters, cv=5, return_train_score=False)
clf.fit(x_train, y_train)

# Access the results using 'cv_results_'
results = clf.cv_results_
print(results)
```

```python
df=pd.DataFrame(results)
df
```

```python
df[['param_C' , 'param_kernel','mean_test_score']]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'param': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'param': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'param': {
            'C': [1, 5, 10]
        }
    }
}

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['param'], cv=5, return_train_score=False)
    clf.fit(x_train, y_train)
    print(f"Best parameters for {model_name}: {clf.best_params_}")
    print(f"Best score for {model_name}: {clf.best_score_}\n")
```
