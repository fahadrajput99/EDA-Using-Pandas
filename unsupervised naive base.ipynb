```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic (1).csv')
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Embarked','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
df['Sex']=la.fit_transform(df['Sex'])
```

```python
df
```

```python
target=df.Survived
input_data = df.drop('Survived', axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
GN=GaussianNB()
```

```python
GN.fit(x_train,y_train)
```

```python
GN.score(x_test,y_test)
```
