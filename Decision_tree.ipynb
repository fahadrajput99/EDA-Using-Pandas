```python
import pandas as pd
```

```python
df=pd.read_csv('/content/salaries.csv')
df
```

```python
input=df.drop(['salary_more_then_100k'],axis=1,inplace=True)
input=df
```

```python
target=df['salary_more_then_100k']
```

```python
input
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
input['compay_n']=la.fit_transform(input['company'])
input['job_n']=la.fit_transform(input['job'])
input['degree_n']=la.fit_transform(input['degree'])
```

```python
input
```

```python
input.drop(columns=['company', 'job', 'degree'], inplace=True)
```

```python
input
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.1)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input,target)
```

```python
model.predict(x_test)
```

```python
model.score(input,target)*100
```
