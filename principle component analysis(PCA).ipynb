```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder
```

```python
df.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis=1 , inplace=True)
```

```python
y=df['HeartDisease']
x=df.drop(['HeartDisease'] , axis=1 , inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
la=LogisticRegression(max_iter=1000)
```

```python
la.fit(X_train , y_train)
```

```python
la.score(X_test , y_test)
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train , y_train)
svm.score(X_test , y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train , y_train)
rfc.score(X_test , y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(df)
x_pca.shape
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
from sklearn.model_selection import train_test_split
X_train_pca, X_test_pca, y_train, y_test = train_test_split(x_pca,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
la=LogisticRegression(max_iter=1000)
la.fit(X_train_pca , y_train)
la.score(X_test_pca , y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=20)
rfc.fit(X_train_pca , y_train)
rfc.score(X_test_pca , y_test)
```
