```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df=pd.read_csv('/content/homeprices.csv')
df
```

```python
import math
beadroom_median=math.floor(df.bedrooms.median())
beadroom_median
```

```python
df.bedrooms=df.bedrooms.fillna(beadroom_median)
```

```python
df
```

```python
import sklearn.linear_model as linear_model

reg = linear_model.LinearRegression()  # Instantiate the LinearRegression class
reg.fit(df[['area', 'bedrooms', 'age']], df['price'])  # Fit the model with the specified features and target variable
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
