```python
import pandas as pd
df=pd.read_csv('/content/Melbourne_housing_FULL.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Distance' , 'Postcode' , 'Lattitude' , 'Longtitude'],axis=1 ,inplace = True)
```

```python
df_fill_zero=['Bedroom2' , 'Bathroom' ,'Car' , 'Propertycount']
df[df_fill_zero]=df[df_fill_zero].fillna(0)
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.Landsize.mean())
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df , drop_first=True)
df
```

```python
y=df.Price
x=df.drop('Price' , axis=1 , inplace=True)
```

```python
df.isnull().sum()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df,y,test_size=0.2)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test , y_test)
```

```python
from sklearn import linear_model
clf=linear_model.Lasso(alpha=50 , max_iter=1000 , tol=0.1)
clf.fit(x_train , y_train)
```

```python
clf.score(x_test , y_test)
```

```python
clf.score(x_train , y_train)
```

```python
from sklearn.linear_model import Ridge
rig_reg=Ridge(alpha=50 , max_iter=1000 , tol=0.1)
rig_reg.fit(x_train , y_train)
```

```python
rig_reg.score(x_train , y_train)
```
